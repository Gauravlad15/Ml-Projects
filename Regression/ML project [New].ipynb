Outlier Treatment

```python
# Step 1 - Boxplot se dekho
sns.boxplot(x=df['Life Expectancy'])
plt.title('Before Outlier Removal')
plt.show()

# Step 2 - IQR se remove karo
Q1 = df['Life Expectancy'].quantile(0.25)
Q3 = df['Life Expectancy'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Life Expectancy'] >= Q1 - 1.5*IQR) & 
        (df['Life Expectancy'] <= Q3 + 1.5*IQR)]

# Step 3 - Verify
sns.boxplot(x=df['Life Expectancy'])
plt.title('After Outlier Removal')
plt.show()
```

#Business Insigths

## 📊 Business Insights

### 1. GDP & Life Expectancy
Countries with higher GDP have significantly better 
life expectancy. Governments should focus on 
economic growth to improve citizen health.

### 2. Adult Mortality — Most Important Feature
Adult mortality is the strongest predictor of life 
expectancy. Reducing adult deaths through better 
healthcare can significantly increase life expectancy.

### 3. Schooling & Education
Countries with higher education levels show better 
life expectancy. Developing countries should invest 
heavily in education infrastructure.

### 4. Developing vs Developed Countries
- Developed countries → 75+ years life expectancy
- Developing countries → 60-65 years life expectancy
- Key action: Improve healthcare infrastructure in 
  developing nations

### 5. BMI & Lifestyle
Countries with healthier average BMI show better 
life expectancy. Promoting healthy lifestyle and 
reducing obesity can improve national health outcomes.

### 6. Overall Recommendation
Focus on these 3 key areas:
- 💰 Increase GDP & economic development
- 🏥 Reduce adult mortality via better healthcare
- 📚 Invest in education & schooling
